# 2.Examine the data by using visualizations.

```python
plt.figure(figsize=(16,6))
country_data.groupby("Region")['Agricultural production index '].mean().plot(kind = "bar",color="red")
plt.title("Region v/s Agricultural production index ",fontsize = 16)
plt.xticks(rotation = 45,fontsize = 10)
plt.show()
```

From the graph it is clear that South Eastern Asia has highest average agricultural production index.

```python
country_data.groupby('Region')['GDP per capita (current US$)'].sum().plot(kind='bar')
plt.title('Highest GDP with respect to region')
plt.show()
```

Northern Europe has highest GDP

```python
country_data.groupby(by=['Region'])['International trade: Imports (million US$)'].mean().plot(kind='bar')
plt.title('Region vs International trade')
plt.show()
```

North America has the highest average Internation trade
